```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('2006_2017_features.csv')
```

```python
df.loc[:,'date'] = pd.to_datetime(df['date'])
```

```python
df.set_index('date', inplace=True)
```

```python
df['day'] = df.index.day
```

```python
df.shape
```

```python
df.reset_index(inplace=True)
```

```python
df.head(15)
```

```python
# cumulative number in the past 10 years
blue = []
red = []
green = []
for i in range(1,1828):
    blue.append(df['color'].iloc[:i].value_counts()['blue'])
    
for i in range(2,1828):
    red.append(df['color'].iloc[:i].value_counts()['red'])
    
for i in range(3,1828):
    green.append(df['color'].iloc[:i].value_counts()['green'])

blue = [0] + blue
red = [0,0] + red
green = [0,0,0] + green
```

```python
df['red'] = red[0:1827]
df['green'] = green[0:1827]
df['blue'] = blue[0:1827]
```

```python
df.head()
```

```python
# cumulative number in the past month
blue_26 = []
red_26 = []
green_26 = []

for i in range(26,1828):
    blue_26.append(df['color'].iloc[i-25:i].value_counts()['blue'])

for i in range(26,1828):
    red_26.append(df['color'].iloc[i-25:i].value_counts()['red'])
    
for i in range(26,1828):
    green_26.append(df['color'].iloc[i-25:i].value_counts()['green'])

blue_26 = [0]*26 + blue_26
red_26 = [0]*26 + red_26
green_26 = [0]*26 + green_26
```

```python
df['red_26'] = red_26[0:1827]
df['green_26'] = green_26[0:1827]
df['blue_26'] = blue_26[0:1827]
```

```python
df.tail(20)
```

```python
df_clean = df[['year','color','red','green','blue','index']]
```

```python
df_clean.head()
```

```python
dict_color = {'red':1, 'green':2, 'blue':3}
color = df_clean['color'].map(dict_color)
df_clean.loc[:,'color'] = color
```

```python
# holdout = df_clean[df_clean['year']==2017]
# df_clean = df_clean[df_clean['year']!=2017]
df_clean = df_clean.drop('year', axis=1)
```

```python
df_clean.tail()
```

```python
X = df_clean.drop('color', axis=1)
y = df_clean['color']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
clf_log = LogisticRegression()
clf_svc = SVC()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
clf_log.fit(X_train, y_train)
clf_svc.fit(X_train, y_train)
```

```python
score = cross_val_score(clf_log, X_train, y_train, cv=5)
score_svc = cross_val_score(clf_svc, X_train, y_train, cv=5)
```

```python
print(score)
print(score_svc)
```

```python
pred_log = clf_log.predict(X_test)
pred_svc = clf_svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(pred_log, y_test))
print('\n')
print(classification_report(pred_log, y_test))
```

```python
print(confusion_matrix(pred_svc, y_test))
print('\n')
print(classification_report(pred_svc, y_test))
```
